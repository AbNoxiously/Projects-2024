```python
df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis="columns", inplace=True)
```

```python
object_col = []
for column in df.columns:
    if df[column].dtype == object and len(df[column].unique()) <= 30:
        object_col.append(column)
        print(f"{column} : {df[column].unique()}")
        print(df[column].value_counts())
        print("====================================")
object_col.remove('Attrition')
```

```python
len(object_col)
```

```python
label = LabelEncoder()
df["Attrition"] = label.fit_transform(df.Attrition)
```

```python
disc_col = []
for column in df.columns:
    if df[column].dtypes != object and df[column].nunique() < 30:
        print(f"{column} : {df[column].unique()}")
        disc_col.append(column)
        print("====================================")
disc_col.remove('Attrition')
```

```python
cont_col = []
for column in df.columns:
    if df[column].dtypes != object and df[column].nunique() > 30:
        print(f"{column} : Minimum: {df[column].min()}, Maximum: {df[column].max()}")
        cont_col.append(column)
        print("====================================")
```

```python
sns.set(style="whitegrid", palette="pastel")

department_counts = df['Department'].value_counts()

department_colors = {'Research & Development': 'skyblue', 'Sales': 'lightgreen', 'Human Resources': 'lightcoral'}

plt.figure(figsize=(8, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', colors=[department_colors[dept] for dept in department_counts.index])
plt.title('Population Distribution by Department')

plt.legend(title='Department', labels=department_colors.keys(), loc='upper right')

plt.show()

fl.fig_caption("A Pie Chart of the Employee Population per Department",
               "A Pie Chart generated from the dataset for Exploratory Data Analysis purposes")
```

```python
sns.set(style="whitegrid", palette="pastel")

legend_labels = {0: 'No', 1: 'Yes'}

plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='MonthlyIncome', hue='Attrition', multiple="stack", kde=False, bins=30, palette={0: 'lightcoral', 1: 'lightgreen'})
plt.title('Distribution of MonthlyIncome by Attrition')
plt.xlabel('MonthlyIncome')
plt.ylabel('Count')
plt.legend(title='Attrition', labels=[legend_labels[label] for label in df['Attrition'].unique()])
plt.show()

fl.fig_caption("A Stacked Bar Plot on the Distribution of Monthly Income by Attrition",
               "A stacked Bar Plot Generated from the dataset for Exploratory Data Analysis")
```

```python
df['Attrition'] = df['Attrition'].astype('category')

sns.set(style="whitegrid", palette="pastel")

plt.figure(figsize=(12, 6))

legend_labels = {0: 'Yes', 1: 'No'}

sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', multiple="stack", kde=False, bins=30, palette={0: 'lightgreen', 1: 'lightcoral'})

plt.legend(title='Attrition', labels=[legend_labels[label] for label in df['Attrition'].cat.categories])

plt.title('Distribution of YearsAtCompany by Attrition')
plt.xlabel('YearsAtCompany')
plt.ylabel('Count')
plt.show()

fl.fig_caption("A Stacked Bar Plot on the Distribution of Years At Company by Attrition",
               "A stacked Bar Plot Illustritaing Distribution of Years  At Company By Attrition Generated from the dataset for Exploratory Data Analysis")
```

```python
df['Attrition'] = df['Attrition'].map({0: 'No', 1: 'Yes'})

sns.set(style="whitegrid", palette="pastel")

plt.figure(figsize=(12, 6))
sns.countplot(x='Department', hue='Attrition', data=df, palette={'No': 'lightcoral', 'Yes': 'lightgreen'})
plt.title('Population Distribution per Department with respect to Attrition')
plt.xlabel('Department')
plt.ylabel('Count')
plt.legend(title='Attrition', labels=['Yes', 'No'])
plt.show()

fl.fig_caption("A Double Bar Graph on the Population Distribution per Department with Respect to Attrition",
               "A Double Bar Graph illustrating the Population Distribution per Department with Respect to Attrition Generated from the dataset for Exploratory Data Analysis Puposes")
```

```python
sns.set(style="whitegrid", palette="pastel")

plt.figure(figsize=(16, 8))
sns.boxplot(data=df, x='JobRole', y='MonthlyIncome', palette='pastel')
plt.title('MonthlyIncome by JobRole')
plt.xlabel('JobRole')
plt.ylabel('MonthlyIncome')
plt.xticks(rotation=45, ha='right') 
plt.show()

fl.fig_caption("A BoxPlot on the Distribution of Monthly Income by Job Role",
               "A BoxPlot Illustrating the Distribution of Monthly Income by Job Role generated from the datastet for Exploratory Data Analysis Purposes")
```

```python
sns.set(style="whitegrid", palette="pastel")

plt.figure(figsize=(12, 8))
sns.countplot(y='JobRole', hue='Attrition', data=df, palette={'Yes': 'lightcoral', 'No': 'lightgreen'})
plt.title('Attrition Counts by Job Role')
plt.xlabel('Count')
plt.ylabel('Job Role')
plt.legend(title='Attrition', labels=['No', 'Yes'])
plt.show()

fl.fig_caption("A Horizontal Bar Graph on the Attrition Counts By Job Role",
               "A Horizontal Bar Graph Ilustraing the Attrition Counts by Job Role generated from the dataset for Exploratory Data Analysis Purposes")
```
